```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_seeds= pd.read_csv(r'C:\Users\lenovo\Documents\GitHub\MSIT_ML_CLASS\datasets\seeds.csv')
```

```python
data_seeds.head()
```

```python
data_seeds = np.array(data_seeds)
```

```python
data_seeds
```

```python
x,y = data_seeds[:,:-1], data_seeds[:,-1]
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3)
model.fit(x)
label = model.predict(x)
print(label)
```

```python
actual = data_seeds[:,-1]
```

```python
df = pd.DataFrame({'predictions':label,'actual':actual})

print(df)

pd.crosstab(df['predictions'],df['actual'])
```

```python
model.inertia_
```

```python
plt.scatter(x[:,1],x[:,2],c = label)
```

```python
plt.scatter(x[:,1],x[:,2], c = y)
```

```python
n_cluster = np.arange(1,10)

inertias = []
for cluster in n_cluster:
    model = KMeans(n_clusters= cluster)
    model.fit(x)
    inertias.append(model.inertia_)
    
for num, inertia in zip(n_cluster, inertias):
    print(num, inertia)
```

```python
plt.plot(n_cluster, inertias, '-o')
plt.xlabel('# of cluster')
plt.ylabel('Inertia')
plt.title('Inertia plot for Seeds Dataset')
plt.show()
```
