```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
data_stock = pd.read_csv(r"C:\Users\lenovo\Documents\GitHub\MSIT_ML_CLASS\datasets\stocks.txt")
```

```python
data_stock.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
len(data_stock['Unnamed: 0'].unique())
```

```python
le = LabelEncoder()

le.fit(data_stock['Unnamed: 0'])
data_stock['Unnamed: 0'] = le.transform(data_stock['Unnamed: 0'])
```

```python
data_stock.head()
```

```python
x,y = data_stock.iloc[:,1:], data_stock.iloc[:,0]

clus = KMeans(n_clusters = 60)

clus.fit(x)
pred = clus.predict(x)

pred
```

```python
df = pd.DataFrame({'predictions':pred,'actual':data_stock.iloc[:,0]})

pd.crosstab(df['predictions'],df['actual'])
```
