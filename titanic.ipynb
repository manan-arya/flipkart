# Applying algorithms after using 'mean' directly for imputer

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_test_ml = pd.read_csv(r"C:\Users\lenovo\Documents\GitHub\MSIT_ML_CLASS\datasets\titanic_test.csv")
```

```python
data_test_ml.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import StandardScaler
```

```python
data_ml.replace('male',0, inplace = True)
data_ml.replace('female',1,inplace = True)

data_ml.head()
```

```python
X = data_ml.iloc[:,1:]
y = data_ml.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 15)
```

```python
X_train.head()
```

```python
steps = [('imp',Imputer(missing_values = "NaN",strategy = 'mean',axis = 0)),('sclae',StandardScaler()),('clf',KNeighborsClassifier(n_neighbors = 6))]
```

## knn

```python
pipeline = Pipeline(steps)

pipeline.fit(X_train,y_train)
pipeline.predict(X_test)
pipeline.score(X_test,y_test)
```

```python
print(pipeline.predict(X_test))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
steps_dt = [('imp',Imputer(missing_values = "NaN",strategy = 'mean',axis = 0)),('sclae',StandardScaler()),('dt',DecisionTreeClassifier())]
```

```python
pipeline_dt = Pipeline(steps_dt)
pipeline_dt.fit(X_train,y_train)
pipeline_dt.score(X_test,y_test)
```

## XGBoost

```python
import xgboost as xgb
```

```python
steps_xgb = [('imp',Imputer(missing_values = "NaN",strategy = 'mean',axis = 0)),('sclae',StandardScaler()),('xgb_clf',xgb.XGBClassifier(objective='binary:logistic', n_estimators=10, seed=123))]
```

```python
from sklearn.model_selection import GridSearchCV

xgb_clf = Pipeline(steps_xgb)

clf_param_grid = {
    'xgb_clf__max_depth' : [2,3,4],
}


estimator = GridSearchCV(xgb_clf, param_grid=clf_param_grid ,cv =4, verbose=1)
estimator.fit(X_test,y_test)
estimator.score(X_test,y_test)

#print(estimator.best_params_)
```

# Applying algorithms after using groupby and then imputing values for age

```python
data_titanic.head()
```

```python
salutation_list = []
for row in range(len(data_titanic)):
    salutation_list.append(data_titanic.iloc[row,3].split(',')[1].split('.')[0])
```

```python
data_titanic['Salutation'] = salutation_list
```

```python
data_titanic["low_fare"] = data_titanic["Fare"]<=32
data_titanic["high_fare"] = data_titanic["Fare"] > 32
data_titanic.head(10)
```

```python
data_titanic.Salutation.unique()
```

```python
data_titanic.groupby('Salutation')['Age'].mean()
```

```python
from sklearn.preprocessing import Imputer
```

```python
data_titanic["Age"] = data_titanic.groupby("Salutation").transform(lambda x: x.fillna(x.mean()))
```

```python
data_titanic.isnull().sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
data_titanic.head()
```

```python
data_titanic.drop("Name", axis = 1, inplace = True)
data_titanic.drop("Embarked", axis = 1, inplace = True)
```

```python
data_titanic.drop("Ticket", axis = 1, inplace = True)
```

```python
data_titanic.replace('male',0,inplace = True)
data_titanic.replace('female',1,inplace = True)
```

```python
data_titanic.head()
```

## knn

```python
le = LabelEncoder()
le.fit(data_titanic['Salutation'])
data_titanic['Salutation'] = le.transform(data_titanic['Salutation'])

steps = [('imp', Imputer(missing_values = 'NaN',strategy = 'mean', axis = 0)),('sclae',StandardScaler()),('knn',KNeighborsClassifier(n_neighbors = 9))]

param_grid = {}

X = data_titanic.iloc[:,2:]
y = data_titanic.iloc[:,1]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)

pipeline = Pipeline(steps)

pipeline.fit(X_train,y_train)
pipeline.predict(X_test)
pipeline.score(X_test,y_test)
```

## Decision Tree

```python
steps_dt = [('imp',Imputer(missing_values = "NaN",strategy = 'mean',axis = 0)),('sclae',StandardScaler()),('dt',DecisionTreeClassifier())]
```

```python
pipeline_dt = Pipeline(steps_dt)
pipeline_dt.fit(X_train,y_train)
pipeline_dt.score(X_test,y_test)
```

## XGBoost

```python
steps_xgb = [('imp',Imputer(missing_values = "NaN",strategy = 'mean',axis = 0)),('sclae',StandardScaler()),('xgb_clf',xgb.XGBClassifier(objective='binary:logistic', n_estimators=10, seed=123))]

xgb_clf = Pipeline(steps_xgb)

clf_param_grid = {
    'xgb_clf__max_depth' : [2,3,4,5],
    'xgb_clf__n_estimators' : [20,30]
}


estimator = GridSearchCV(xgb_clf, param_grid=clf_param_grid ,cv =4, verbose=1)
estimator.fit(X_test,y_test)
estimator.score(X_test,y_test)
```
